```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = 2*x + 3

plt.plot(x, y, color='blue', label='y = 2x + 3')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Doğrusal Fonksiyon')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y2 = x**2 - 4

plt.plot(x, y2, color='red', label='y = x² - 4')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Kuadratik Fonksiyon')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y3 = np.sin(x)

plt.plot(x, y3, color='green', label='y = sin(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Sinüs Fonksiyonu')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(x, y, color='blue', label='y = 2x + 3')
axes[0].set_title('Doğrusal')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(x, y2, color='red', label='y = x² - 4')
axes[1].set_title('Kuadratik')
axes[1].legend()
axes[1].grid(True)

axes[2].plot(x, y3, color='green', label='y = sin(x)')
axes[2].set_title('Sinüs')
axes[2].legend()
axes[2].grid(True)

plt.suptitle('Fonksiyon Grafikleri', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Veri seti simülasyonu (gerçek veri gibi)
np.random.seed(42)
gunler = np.arange(1, 31)
satislar = 50 + 3*gunler + np.random.randn(30)*10

plt.figure(figsize=(10, 5))
plt.scatter(gunler, satislar, color='orange', label='Günlük Satış', zorder=5)
plt.plot(gunler, 50 + 3*gunler, color='blue', linestyle='--', label='Trend (y = 50 + 3x)')
plt.title('30 Günlük Satış Verisi')
plt.xlabel('Gün')
plt.ylabel('Satış Adedi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Doğrusal
axes[0,0].plot(x, y, 'b-', label='y = 2x + 3')
axes[0,0].set_title('Doğrusal Fonksiyon')
axes[0,0].grid(True)
axes[0,0].legend()

# Kuadratik
axes[0,1].plot(x, y2, 'r-', label='y = x² - 4')
axes[0,1].set_title('Kuadratik Fonksiyon')
axes[0,1].grid(True)
axes[0,1].legend()

# Sinüs
axes[1,0].plot(x, y3, 'g-', label='y = sin(x)')
axes[1,0].set_title('Sinüs Fonksiyonu')
axes[1,0].grid(True)
axes[1,0].legend()

# Satış verisi
axes[1,1].scatter(gunler, satislar, color='orange', label='Satış')
axes[1,1].plot(gunler, 50 + 3*gunler, 'b--', label='Trend')
axes[1,1].set_title('Satış Verisi')
axes[1,1].grid(True)
axes[1,1].legend()

plt.suptitle('Cebir - Fonksiyon Grafikleri', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
